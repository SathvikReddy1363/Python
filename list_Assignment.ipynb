```python
#Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.
import random
random_lst = [random.randint(1, 100) for _ in range(20)]
print("Original list:", random_lst)
aclst = sorted(random_lst)
print("Sorted list (ascending):", aclst)
dsc_lst = sorted(random_lst, reverse=True)
print("Sorted list (descending):", dsc_lst)
unique_lst = list(set(random_lst))
print("List with duplicates removed:", unique_lst)
```

```python
#Create a list of dictionaries where each dictionary represents a student with keys 'name' and 'score'. Sort the list of dictionaries by the 'score' in descending order and print the sorted list.
students = [ {"name":"Sathvik", "score":99},
            {"name":"Siri", "score":95},
            {"name":"Lakshmi", "score":100},
            {"name":"Rohit", "score":98},
            {"name":"Ananya", "score":97}
]
Sorted_lst = sorted(students, key=lambda x: x['score'], reverse=True)
#print("Sorted list of students by score (descending):", Sorted_lst)
for student in Sorted_lst:
    print(f"Name: {student['name']}, Score: {student['score']}")
```

```python
#Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.
def transpose(matrix):
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]

# Test the function
original_matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
transposed_matrix = transpose(original_matrix)

print("Original Matrix:")
for row in original_matrix:
    print(row)

print("\nTransposed Matrix:")
for row in transposed_matrix:
    print(row)
```

```python
#Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.
def flatten(nested_list):
    return [item for sublist in nested_list for item in sublist]
# Test the function
nested_list = [[1, 2], [1, 4], [5,  6]]
flattened_list = flatten(nested_list)
print("Original Nested List:", nested_list)
print("Flattened List:", flattened_list)
```

```python
#Create a list of the first 10 positive integers. Remove the elements at indices 2, 4, and 6, and insert the element '99' at index 5. Print the modified list.
numbers = [i for i in range(1,11)]
print("Original list:", numbers)
new_list = [num for index, num in enumerate(numbers) if not index in[2, 4, 6]]
new_list.insert(5,99)
print("Modified list:", new_list)
```

```python
#Create two lists of the same length. Use the `zip` function to combine these lists into a list of tuples and print the result.
list1 = ['a', 'b', 'c', 'd', 'e']
list2 = [1, 2, 3, 4, 5,6]
zip_list = list(zip(list1, list2))
print("Zipped List:", zip_list)
```

```python
#Write a function that takes a list and returns a new list with the elements in reverse order. Print the original and reversed lists.
def reverse(lst):
    return lst[::-1]
lst = [1, 2, 3, 4, 5]
reversed_lst = reverse(lst)
print("Original List:", lst)
print("Reversed List:", reversed_lst)
```

```python
#Write a function that rotates a list by n positions. Print the original and rotated lists.
def rotate(lst, n):
    #n = n % len(lst)  # Handle cases where n is greater than the list length
    return lst[n:n+1] + lst[n+1:] + lst[:n]

lst = [1, 2, 3, 4, 5]
rotated_lst = rotate(lst, 2)
print("Original List:", lst)
print("Rotated List:", rotated_lst)
```

```python
#Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.
def intersect(list1, list2):
    return list(set(list1) & set(list2))

list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
intersected_list = intersect(list1, list2)
print("List 1:", list1)
print("List 2:", list2)
print("Intersected List:", intersected_list)
```
